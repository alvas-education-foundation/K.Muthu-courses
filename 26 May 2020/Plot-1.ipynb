```python
#Python program to demonstrate various plotting types using subplots


#importing matplotlib module
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# x-axis values
x=[1,2,3,4,5]
#y-axis values
y=[2,6,4,8,3]
```

```python
fig=plt.figure()
```

```python
#Creating 2*2 form of subplots
fig , axes = plt.subplots(2,2)
axes[(0,0)].plot(x,y)
axes[(0,1)].bar(x,y)
axes[(1,0)].scatter(x,y)
axes[(1,1)].fill(x,y)
plt.tight_layout()
```
