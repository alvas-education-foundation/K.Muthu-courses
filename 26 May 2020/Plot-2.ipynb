```python
#Python program to demonstrate linestyle ans marker attribute in plotting

#importing matplotlib module
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#x-axis values
x=[1,2,3,4,5]
#y-axis values
y=[2,3,4,7,9]
```

```python
fig1 = plt.figure()
```

```python
fig1 , axes = plt.subplots(1,2)
#plotting graph with line style
axes[0].plot(x,y,color='g',linestyle='--')
#plotting graph with marker
axes[1].plot(x,y,color='r',marker='o')
plt.tight_layout()
```
