```python
#Python program to locate all volcanoes in the world
import numpy as np
import pandas as pd
import folium as fo
%matplotlib inline
```

```python
map=fo.Map()
```

```python
volcano=pd.read_csv("volcano.csv")
```

```python
volcano.head()
```

```python
name_vo=list(volcano["Name"])
lat_vo=list(volcano["Latitude"])
lon_vo=list(volcano["Longitude"])
```

```python
vo=fo.FeatureGroup(name="My Map")
```

```python
for name,lat,lon in zip(name_vo,lat_vo,lon_vo):
    vo.add_child(fo.Marker(location=[lat,lon],popup=name,icon=fo.Icon(color="red")))
```

```python
map.add_child(vo)
```
