```python
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flight2 = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
flight2
```

```python
sns.heatmap(flight2,annot=True,lw=1)
```

```python
sns.clustermap(flight2,cmap='cool')
```
